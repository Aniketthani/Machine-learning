In this part of this exercise, you will implement linear regression with one
variable to predict profits for a food truck. Suppose you are the CEO of a
restaurant franchise and are considering different cities for opening a new
outlet. The chain already has trucks in various cities and you have data for
profits and populations from the cities.

You would like to use this data to help you select which city to expand
to next.
The file ex1data1.txt contains the dataset for our linear regression prob-
lem. The first column is the population of a city and the second column is
the profit of a food truck in that city. A negative value for profit indicates a
loss

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Loading Data

```python
data=pd.read_csv('data/ex1data1.txt')
```

```python
data.head()
```

```python
X=data['Population']
```

```python
Y=data['Profit']
```

```python
m=len(Y) #no. of training examples
```

# Plotting the Data

```python
plt.plot(X,Y,'rx')
plt.ylabel("Profit in $10,000s")
plt.xlabel("Population of City in 10,000s");
```

# Initialize Theta

```python
theta=np.zeros((2,1))
```

```python
theta.shape
```

# Adding X0=1 in X

```python
s=pd.Series([1 for i in range(m)])
X=pd.concat([s,X],axis=1)
X.columns=['x0','Population']
X
```

```python
X=X.to_numpy()
Y=Y.to_numpy()
Y=Y.reshape((97,1))
```

# Cost Function

```python
def ComputeCost(X,Y,theta):
    m=X.shape[0]
    return (0.5/m)*np.sum((np.dot(X,theta) -Y)**2)
    
```

```python
print(ComputeCost(X,Y,theta))
```

# Gradient Descent 

```python
def GradientDescent(X,Y,theta,alpha,iterations):
    
    m=Y.shape[0]
    
    J_history=np.zeros((iterations,1))
    
    for i in range(iterations):
        theta=theta - (alpha/m)*(np.dot(X.T,(np.dot(X,theta)-Y)))
        J_history[i]=ComputeCost(X,Y,theta)
        print(J_history[i])
        
    return theta
```

```python
theta=GradientDescent(X,Y,theta,0.01,1500)
```

```python
theta
```

# Hypothesis function/Prediction function

```python
def hypothesis(x,theta):
    return (np.dot(x,theta))
```

# Visualization of Hypothesis

```python
plt.plot(X[:,1],Y,'rx')
predictions=hypothesis(x,theta)
plt.plot(x[:,1],predictions,'b');
```
